```python
#Corregir los target a solo dos valores distintos â€˜<=50kâ€™ y â€˜>50kâ€™
df_wdata = df_inicialwdata.Target.replace({" <=50K.":" <=50K"," >50K.":" >50K"})
df_wdata
```

Calcular la proporciÃ³n que hay entre personas con salarios >50k respecto del total para 
Entre personas de distinto sexo, Raza y Educacion. Por ejemplo cuÃ¡l es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y EducaciÃ³n


```python
df_unificado.Sex.unique()#busco saber cuales son los sexos que hay en total
```

```python
df_unificado.Race.unique()#busco saber cuales son las razas que hay en total
```

```python
df_unificado.Education.unique()#busco saber cuales son la educacion que hay en total
```

```python
mayor_salario = df_unificado[df_unificado.Target.isin([' >50K'])]
mayor_salario
```

```python
mayor_salario.dropna()
```

```python
x_genero = mayor_salario.groupby('Sex').agg({'Target':'count'}).sort_values('Target', ascending=False)
x_genero
```

```python
x_genero['percent'] = (x_genero['Target'] / x_genero['Target'].sum()) * 100
x_genero
```

```python
x_raza = mayor_salario.groupby('Race').agg({'Target':'count'}).sort_values('Target', ascending=False)
x_raza
```

```python
x_raza['percent'] = (x_raza['Target'] / x_raza['Target'].sum()) * 100
x_raza
```

```python
x_educacion = mayor_salario.groupby('Education').agg({'Target':'count'}).sort_values('Target', ascending=False)
x_educacion
```

```python
x_educacion['percent'] = (x_educacion['Target'] / x_educacion['Target'].sum()) * 100
x_educacion
```
