Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

```python
df_inicial.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,ax = plt.subplots(figsize=(20,4))
ax.hist(df_inicial['Salary'],bins=100)  
ax.set_xlim(df_inicial['Salary'].min(),df_inicial['Salary'].max())

ax.set_title("Distribución de Salarios")

plt.show()
```

```python
df_inicial['Salary'].mean()
```

```python
fig,ax = plt.subplots(figsize=(20,8))
sns.histplot(df_inicial,x='Salary', hue='GenderID', multiple='stack', palette='magma')
ax.set_title("Distribución de Salarios")
```

Realizar un gráfico de violín con los salarios, discriminados por género. 

```python
fig2,ax = plt.subplots(figsize=(10,10))
sns.violinplot(data=df_inicial, y='Salary', x='GenderID', palette='magma')
ax.set_title("Distribución de Salarios")

plt.show()
```

```python
#Analizo sumando RaceDesc

fig2,ax = plt.subplots(figsize=(10,10))
sns.violinplot(data=df_inicial, x='RaceDesc', y='Salary', hue='GenderID', split=True, palette='magma')
ax.set_title("Distribución de Salarios")
ax.tick_params(axis = 'x',labelsize=8,rotation=45)

plt.show()
```

Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire)


```python
df_inicial.head()
```

```python
df_inicial[['DateofHire','Salary']]
```

```python
df = df_inicial[['DateofHire','Salary']]
df
```

```python
df.dtypes
```

```python
df['DateofHireDT'] = pd.to_datetime(df['DateofHire']) #Transformo el año valor float a datetime
df.dtypes
```

```python
df
```

```python
df_final = df.drop(columns=['DateofHire']).copy()
df_final.set_index('DateofHireDT', inplace=True)
df_final
```

```python
df_final.groupby(by='DateofHireDT').mean()
```

```python
fig3,ax = plt.subplots(figsize=(30,8))
ax=sns.lineplot(data=df_final, x='DateofHireDT', y='Salary', sort=True , ci= None )
ax.tick_params(axis = 'x',labelsize=12)
```
