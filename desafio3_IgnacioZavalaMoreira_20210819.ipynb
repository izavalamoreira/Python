```python
df_pais_medalla = df_olimpiadas_limpio.groupby(by='Country',as_index=False).count()[['Country','Medal']] # Genero nuevo df agrupando por País. Agrupo contando porque cada fila representa una medalla
df_pais_medalla
```

```python
df_pais_medalla.loc[df_pais_medalla['Medal'] == df_pais_medalla['Medal'].max()] # encuentro el máximo loqueando el que coincide con el máximo
```
