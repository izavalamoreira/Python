# Punto 1

Construir un objeto de series de tiempo con el índice igual al año de los juegos olímpicos

```python
df_clean.head()
```

```python
df_clean['Year_Aux'] = pd.to_datetime(df_clean['Year'], format = '%Y') #Transformo el año valor float a datetime
```

```python
df_clean.head()
```

```python
df_anio = df_clean[['City','Year_Aux']]
df_anio
```

```python
df_anio.drop_duplicates(inplace=True) #Me quedo con una sola fila por año (porque las cols con las que me quedé son iguales para cada año)
df_anio
```

```python
df_anio.rename(columns={'Year_Aux':'Anio'}, inplace=True)
df_anio.set_index('Anio', inplace=True)
df_anio
```

```python
type(df_anio)
```

```python
df_anio_serie = df_anio.squeeze() # Uso squeeze para transformarlo en objeto tipo serie
df_anio_serie
```

# Punto 3

Obtener medidas resumen del conjunto de datos. ¿Cuál es el país que ganó más medallas?

```python
df_clean.drop(columns=['index'], inplace=True)
df_clean
```

```python
# Cantidad de deportes
df_clean['Sport'].nunique()
```

```python
#Primeras Olimpíadas registradas en este df
df_clean['Year'].min()
```

```python
df_ispa = df_clean.groupby(by='Country',as_index=False).count()[['Country','Medal']] # Genero nuevo df agrupando por País. Agrupo contando porque cada fila representa una medalla
df_ispa
```

```python
df_ispa.loc[df_ispa['Medal'] == df_ispa['Medal'].max()] # encuentro el máximo loqueando el que coincide con el máximo
```
